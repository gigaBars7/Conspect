```python
import cv2
import easyocr
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
reader = easyocr.Reader(['ru', 'en'], gpu=True)
```

```python
test_image = 'tests/_.png'
img_rgb = cv2.imread(test_image, cv2.IMREAD_COLOR_RGB)

result = reader.readtext(img_rgb, detail=0, paragraph=True')

plt.imshow(img_rgb)
plt.axis("off")
plt.show()

text = "\n".join(result)
print(text)
```
